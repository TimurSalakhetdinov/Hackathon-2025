```python
from icrawler.builtin import GoogleImageCrawler, BingImageCrawler
import os

def download_broken_tile_images(
    engine="google", 
    keyword="broken roof tile", 
    max_num=200, 
    save_dir="external_data/broken_tiles", 
    min_size=(300, 300)
):
    os.makedirs(save_dir, exist_ok=True)

    # Choose crawler engine
    if engine == "google":
        crawler = GoogleImageCrawler(storage={'root_dir': save_dir})
    elif engine == "bing":
        crawler = BingImageCrawler(storage={'root_dir': save_dir})
    else:
        raise ValueError("engine must be 'google' or 'bing'")

    # Start downloading
    crawler.crawl(
        keyword=keyword, 
        max_num=max_num,
        min_size=min_size,
        file_idx_offset=0
    )

    print(f"✅ Downloaded {max_num} images to {save_dir}")

# Example usage:
download_broken_tile_images(
    engine="google", 
    keyword="broken roof tile close-up", 
    max_num=250
)
```
